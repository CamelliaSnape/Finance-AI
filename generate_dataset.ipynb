```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("personal_finance_dataset_8000_extended.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()

df = df.dropna()
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Amount"] = pd.to_numeric(df["Amount"])
```

```python

df["Category"] = df["Category"].astype(str).str.strip().str.title()
df["TransactionType"] = df["TransactionType"].astype(str).str.strip()
```

```python
print(df.columns.tolist())
```

```python
df["Year"] = df["Date"].dt.year

df["Month"] = df["Date"].dt.month

df["Day"] = df["Date"].dt.day

df["DayOfWeek"] = df["Date"].dt.day_name()
```

```python
encoder = LabelEncoder()

df["Category"] = encoder.fit_transform(df["Category"])

df["TransactionType"] = encoder.fit_transform(df["TransactionType"])
```

```python
df.to_csv("cleaned_finance_dataset.csv", index=False)

print("Đã tạo file cleaned_finance_dataset.csv")
```
